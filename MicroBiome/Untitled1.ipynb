```python
import pandas as pd
# import geopandas
import matplotlib.pyplot as plt
from pylab import *
import pivottablejs as pj
import pandas_profiling as pdf
```

```python
def project_id(link):
    if "=" in link:
        return link.split("=")[-1]
    else:
        return link.rstrip("/").split("/")[-1]
def repo(link):
    return 'mg-rast' if 'mg-rast' in link else 'ncbi'
df = pd.read_csv("/home/devil/Downloads/refined_tot_harmonization - Main.csv")
df["REPO"] = df["REPOSITORY LINK"].apply(repo)
# df["REPOSITORY ID"] = df["REPOSITORY ID"].apply(project_id)
del df["REPOSITORY LINK"]
# del df['SAMPLE NAME ']

for lon_lat in ['31 27','35 -4', '7 9',]:
    df.loc[df['LON LAT']==lon_lat, 'LON LAT'] = lon_lat.replace(" ", ",")
df['LON LAT'] = df['LON LAT'].apply(lambda x: x.replace(" ",""))
df['LON LAT'] = df['LON LAT'].apply(lambda x: ",".join(x.split(",")[:2]))
df[df=="-"] = np.nan

df["LON"] = df['LON LAT'].apply(lambda x: float(x.split(",")[0]) if type(x) ==str else np.nan)
df["LAT"] = df['LON LAT'].apply(lambda x: float(x.split(",")[1]) if type(x) ==str else np.nan)
df = df.rename(columns={'SAMPLE NUMBER': "SAMPLE COUNT", 'ETHNICITY ': 'ETHNICITY',
                     'TARGET AMPLICON ':'TARGET AMPLICON'})
del df['LON LAT']
import re
regex = re.compile('PMC\d+|\d+', flags=re.IGNORECASE)

df['STUDY LINK']= df.groupby(['STUDY LINK']).size().reset_index()["STUDY LINK"].apply(lambda x: regex.findall(x)).apply(','.join)
df.columns = [x.replace(" ","_") for x in df.columns]
# df[['REPOSITORY ID', 'SAMPLE COUNT', 'STUDY TITLE', 'PLATFORM','PLATFORM.1','SAMPLE TYPE']]#.to_csv("samples")
```
